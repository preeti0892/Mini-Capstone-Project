```python
#What is the distribution of victim ages in reported crimes?

#Is there a significant difference in crime rates between male and female victims?

x="select Vict_Age,Vict_Sex,count(*) as crime_count from crime_data group by Vict_Age,Vict_sex"
df1=pd.read_sql(x,connection)
```

```python
sns.scatterplot(data=df1,x="Vict_Age",y="crime_count",hue="Vict_Sex",size="crime_count")
# distribution of victim ages in reported crimes.
```

```python
df["Vict_Sex"].value_counts().reset_index() #maximum victims are male.
```

```python
df["Vict_Sex"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.ylabel("")  # Remove the y-label
```

```python
c=df["Vict_Sex"]=="F"
d=df[c]
h=d["Vict_Age"]
plt.hist(h,bins=8,edgecolor="yellow")
#most of the female victims are of 20 to 30 age group ans their number is above 40.
```

```python
f=df["Vict_Sex"]=="M"
g=df[f]
h=g["Vict_Age"]
plt.hist(h,bins=9,edgecolor="Black")
#most of male victims are between 25 to 35 age group and their number is around 78. 
```

```python
fig , (ax1,ax2)=plt.subplots(1,2,figsize=(12,6))
sns.histplot(data=h,bins=10,edgecolor="red",kde=True,ax=ax1)
ax1.set_title("crime rates in female victims")
ax1.set_xlabel("age")
ax1.set_ylabel("Crimes_Count")

sns.histplot(data=h,bins=10,edgecolor="brown",kde=True,ax=ax2)
ax2.set_title("crime rates in males victims")
ax2.set_xlabel("age")
ax2.set_ylabel("Crimes_Count")
```

```python
sns.histplot(data=df,x=df["Vict_Age"],hue="Vict_Sex",bins=15)
```
